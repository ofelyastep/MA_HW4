```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
from networkx.algorithms.community import greedy_modularity_communities
```

Problem 1

```python

# Creating a directed graph
directed_graph = nx.read_edgelist("connections.txt", create_using=nx.DiGraph(directed=True), nodetype=int)

# Creating an undirected graph
graph = nx.read_edgelist("connections.txt", create_using=nx.Graph(), nodetype=int)
```

Problem 3

The graph has low density, meaning it's very sparse. This implies that only a small number of potential edges exist in the graph, indicating few connections between nodes compared to the total possible connections.

Problem 4

```python
max_degr = max(degr.values())
min_degr = min(degr.values())
```

```python
high_degr_n = [node for node, degree in degr.items() if degree == max_degr]
low_degr_n = [node for node, degree in degr.items() if degree == min_degr]
```

```python
print(f'Highest number of connections: {high_degr_n}') 
print(f'Lowest number of connections: {low_degr_n}')
```

Problem 6

```python
closeness_centr = nx.closeness_centrality(directed_graph)
betweenness_centr = nx.betweenness_centrality(directed_graph)
eigenvector_centr = nx.eigenvector_centrality(directed_graph, max_iter= 500)
```

```python
max_closeness = max(closeness_centr.values())
max_betweenness = max(betweenness_centr.values())
max_eigenvector = max(eigenvector_centr.values())
```

```python
highest_closeness_n = [node for node, centrality in closeness_centr.items() if centrality == max_closeness]
highest_betweenness_n = [node for node, centrality in betweenness_centr.items() if centrality == max_betweenness]
highest_eigenvector_n = [node for node, centrality in eigenvector_centr.items() if centrality == max_eigenvector]
```

```python
print(f'Highest closeness: {highest_closeness_n}, with value {closeness_centr[highest_closeness_n[0]]}') 
print(f'Highest betweenness: {highest_betweenness_n}, with value {betweenness_centr[highest_betweenness_n[0]]}')
print(f'Highest eigenvector: {highest_eigenvector_n}, with value {eigenvector_centr[highest_eigenvector_n[0]]}')
```

Closeness:
Closeness centrality assesses a node's proximity to others in the graph. The node with the top closeness value (0.2809825342704057) maintains the shortest average distance to all other nodes, allowing swift traversal to or from it within the graph. It holds a strategic position for quick interactions with other nodes.

Betweenness:
Betweenness centrality gauges a node's position along the shortest paths between nodes in the graph. The node with the highest betweenness value (0.3377974497301992) significantly influences information flow or interactions within the graph. It governs the flow of information between distinct clusters or groups within the network.

Eigenvector:
Eigenvector centrality quantifies a node's influence by considering the centrality of its neighboring nodes. The node with the top eigenvector value (0.07339141387266628) connects to other nodes that also hold significant centrality in the graph or are well-connected themselves. Its importance stems from being linked to other crucial nodes in the network.

```python
largest_communities = sorted(communities, key=len, reverse=True)[:3]
```

Problem 10

```python
influencers = []
for community in largest_communities:
    subgraph = directed_graph.subgraph(community)
    centrality = nx.degree_centrality(subgraph)
    influencer = max(centrality, key=centrality.get)
    influencers.append(influencer)
```

```python
influencer_subgraphs = []
for influencer in influencers:
    influencer_subgraph = directed_graph.subgraph(nx.ego_graph(directed_graph, influencer, radius=1))
    influencer_subgraphs.append(influencer_subgraph)
```

```python
node_sizes = []
for subgraph in influencer_subgraphs:
    centrality = nx.degree_centrality(subgraph)
    sizes = [80 if node == max(centrality, key=centrality.get) else 30 for node in subgraph.nodes()]
    node_sizes.append(sizes)
```

```python
for i, subgraph in enumerate(influencer_subgraphs):
    plt.figure(i+1)
    node_colors = ['lightblue' if node != influencers[i] else 'red' for node in subgraph.nodes()]
    nx.draw(subgraph, with_labels=False, node_color=node_colors, node_size=node_sizes[i])
    plt.title(f"Top Community {i+1}")
    plt.show()
```


The influential node, marked in red, holds the highest degree centrality among all nodes.
